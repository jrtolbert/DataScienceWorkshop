```python
# define a function that will transform html into digestable list of data  

def parse_table(html_table):
    parsed_table = []
    
    # define size of table by getting a list of the rows within the table
    table_rows = html_table.find_all("tr")
    
    # delete the header rows of the table
    del table_rows[0:3]
    
    # initalize rowspan and colspan variables
    rowspan_list = []
    colspan_list = []

    for i in range(len(table_rows)):
        table_data = table_rows[i].find_all("td")
        print(len(table_data))
        for j in range(len(table_data)):
            if(table_data[j].get('rowspan') != None):
                # add data into rowspan_list
                rowspan_list.append({table_data[j].text.strip('\n'): int(table_data[j].get('rowspan'))})
            elif(table_data[j].get('colspan') != None):
                # add data into colspan_list
                colspan_list.append({table_data[j].text.strip('\n'): int(table_data[j].get('colspan'))})
            else:
                # add remaining data into colspan_list
                colspan_list.append({table_data[j].text.strip('\n'): 1})

    return [rowspan_list, colspan_list]
```
